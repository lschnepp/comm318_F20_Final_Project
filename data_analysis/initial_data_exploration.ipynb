### Transforming the data frame

* One way to work with a data frame like this is to transform it into a long form where you have a row for each county and year combination.


* We can drop the `TOTAL` column and rename the `County` column to `county`

```python
deaths_df2=deaths_df.rename(columns={'County': 'county'})
deaths_df2=deaths_df2.drop(columns=('TOTAL'))
```

* Then we can use the `.melt()` function to reshape the data frame.

```python
deaths_long_df=deaths_df2.melt(id_vars='county', value_name='deaths', var_name='year')
deaths_long_df.shape
```

* There are now 300 rows
  * This is 30 counties * 10 years for each

```python
deaths_long_df.head(20)
```

* Then you can use `.groupby()` to something like this:

```python
deaths_long_df.groupby('county')['deaths'].sum()
```

```python
deaths_by_year=deaths_long_df.groupby(['year','county'])['deaths'].sum().unstack()

deaths_by_year
```
